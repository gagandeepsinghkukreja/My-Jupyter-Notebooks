```python
#Easiest way to assign dummmy variable
#Do not use zero. Modelling values might not retain.
#Give ordinal ranking because training maths will be able to understand the target variable although its not a good practice.

df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55': 6 ,'55+' : 7 })

df
```

```python
#Another_Way_couldbe_label_encoding
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
df['Age']= label_encoder.fit_transform(df['Age'])

df['Age'].unique()
```

```python
#Fixing categorical city_category
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```
